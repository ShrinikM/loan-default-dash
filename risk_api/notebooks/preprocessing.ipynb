```python
keep_cols = [
    # Metadata
    'loan_status', 'issue_d', 'addr_state',
    
    # Borrower Info
    'annual_inc', 'dti', 'emp_length', 'home_ownership', 'verification_status',
    
    # Loan Info
    'loan_amnt', 'term', 'int_rate', 'installment', 'grade', 'purpose', 'application_type',
    
    # Credit History
    'fico_range_low', 'fico_range_high',
    'earliest_cr_line', 
    'open_acc', 'total_acc', 
    'revol_bal', 'revol_util', 
    'pub_rec', 'pub_rec_bankruptcies', 'mort_acc',
    'inq_last_6mths'
]
```

```python
newdf = df1[keep_cols]
```

```python
newdf = newdf.dropna(subset=['loan_status', 'issue_d', 'loan_amnt'])
newdf['mort_acc'] = newdf['mort_acc'].fillna(0)
newdf['pub_rec_bankruptcies'] = newdf['pub_rec_bankruptcies'].fillna(0)


def clean_emp_length(x):
    if pd.isna(x):
        return 0
    x = str(x)
    if '< 1' in x:
        return 0
    if '10+' in x:
        return 10
    return int(''.join(filter(str.isdigit, x)))

newdf['emp_length'] = newdf['emp_length'].apply(clean_emp_length)

newdf['dti'] = newdf['dti'].fillna(newdf['dti'].median())
newdf['revol_util'] = newdf['revol_util'].fillna(newdf['revol_util'].median())

newdf = newdf.dropna()

print(f"Final Row Count: {len(newdf):,}")
print(newdf.isnull().sum().sum())
```

```python
df_clean = newdf[newdf['loan_status'].isin(['Fully Paid', 'Charged Off'])].copy()

df_clean['target'] = df_clean['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)

df_clean['issue_d'] = pd.to_datetime(df_clean['issue_d'], format='%b-%Y')
df_clean['earliest_cr_line'] = pd.to_datetime(df_clean['earliest_cr_line'], format='%b-%Y')


df_clean['credit_hist_len'] = (df_clean['issue_d'] - df_clean['earliest_cr_line']).dt.days / 365.25


df_clean = df_clean.drop(columns=['earliest_cr_line'])

print(f"Original Row Count: {len(newdf):,}")
print(f"Finalized Row Count: {len(df_clean):,}")
print(f"\nTarget Distribution:\n{df_clean['target'].value_counts(normalize=True)}")
```

```python
# Load Unemployment (Monthly)
unemp = pd.read_csv('UNRATE.csv', parse_dates=['observation_date'])
unemp = unemp.rename(columns={'observation_date': 'issue_d', 'UNRATE': 'unemployment_rate'})
unemp = unemp.set_index('issue_d')

# Load Delinquency (Quarterly)
delinq = pd.read_csv('DRCLACBS.csv', parse_dates=['observation_date'])
delinq = delinq.rename(columns={'observation_date': 'issue_d', 'DRCLACBS': 'delinq_rate'})
delinq = delinq.set_index('issue_d')
```

```python
delinq_monthly = delinq.resample('MS').ffill()
macro_data = unemp.join(delinq_monthly, how='outer')
macro_data = macro_data.ffill().reset_index()

macro_data = macro_data[(macro_data['issue_d'] >= '2000-01-01') & (macro_data['issue_d'] <= '2020-01-01')]

print(macro_data.shape)
print(macro_data.head())
```

```python
df_final = df_clean.merge(macro_data, on='issue_d', how='left')

print(df_final.shape)
print(f"Missing Unemployment Values: {df_final['unemployment_rate'].isnull().sum()}")
```
